# Dendrogram with normalization

```python
# Normalization functio 
def norm_func(i):
    x = (i-i.min())//(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame ( considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:])
```

```python
df_norm
```

```python
import scipy.cluster.hierarchy as sch
```

```python
# create dendogram 

dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering
#create cluster 
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean')
```

```python
# save cluster for chart
y_hc = hc.fit_predict(df_norm)
```

```python
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

# K-mean

```python
df3= pd.read_csv(r"C:\Users\shrad\Downloads\crime_data.csv")
df3
```

```python
df3.iloc[:,1:]
```

```python
# Standarization function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df3_df = scaler.fit_transform(df3.iloc[:,1:])
```

```python
scaled_df3_df
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4,random_state=0)
kmeans.fit(scaled_df3_df)
```

```python
df3["Cluster"] = kmeans.predict(scaled_df3_df)
df3
```

# Elbow

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(2,33):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_df3_df)
    dist = kmeans.inertia_
    wcss.append(dist)
```

```python
wcss
```

```python
plt.plot(range(2,33),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of cluster")
plt.ylabel("Within a cluster sum of square")
plt.grid()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_df3_df)
    wcss.append(kmeans.inertia_)
    if i > 6:
      break
```

```python
#Build cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(6, random_state=0)
clusters_new.fit(scaled_df3_df)
```
