```python
#calculating vif


rsq_Profit= smf.ols('Profit~Total_Spend+Profit_Prediction',data=startupN).fit().rsquared
vif_Profit= 1/(1-rsq_Profit)

rsq_Total_Spend= smf.ols('Total_Spend~Profit+Profit_Prediction',data=startupN).fit().rsquared
vif_Total_Spend= 1/(1-rsq_Total_Spend)

rsq_Profit_Prediction= smf.ols('Profit_Prediction~Total_Spend+Profit',data=startupN).fit().rsquared
vif_Profit_Prediction= 1/(1-rsq_Profit_Prediction)
```

```python
#storing vif value in the dataframe 
d1 = { 'Variable':['Profit,Total_Spend,Profit_Prediction'], 'Vif':[vif_Profit,vif_Total_Spend,vif_Profit_Prediction]}
```
