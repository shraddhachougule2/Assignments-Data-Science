```python
import numpy as np
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import warnings
warnings.filterwarnings('ignore')
```

```python
dt = pd.read_csv(r"C:\Users\User\Downloads\wine (1).csv")
dt
```

```python
dt.info()
```

```python
#Removing Unwanted Columns
dt2 = dt.iloc[:,1:]
dt2
```

```python
dt2.describe()
```

```python
#converting into numpy array
dt3 = dt2.values
dt3
```

```python
## numerical data
dt4 = scale(dt3)
dt4
```

# Applyind PCA fit transform to dataset

```python
pca = PCA(n_components=13)
dt_pca = pca.fit_transform(dt4)
dt_pca
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,4)*100)
var1
```

```python
## PCA PLot
plt.plot(var1,color='magenta')
plt.show()
```

```python
dt["Type"]
```

```python
final_data = pd.concat([dt[['Type']],pd.DataFrame(dt_pca[:,0:3],columns=['PC1','PC2','PC3'])], axis = 1)
final_data
```

```python
## Visualisation of PCA's
plt.figure(figsize=(16,12))
sns.scatterplot(data=final_data)
plt.show()
```

# Checking with other Clustering algorithum

```python
## 1. Hierarchical Clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
### Dendogram 
plt.figure(figsize=(16,12))
dendogram = sch.dendrogram(sch.linkage(dt4,'complete'))
plt.show()
```

```python
### Clusters
cl = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
cl
```

```python
y_cl = cl.fit_predict(dt4)
Clusters=pd.DataFrame(y_cl,columns=['Clusters'])
Clusters
```

```python
wine = dt.copy()
wine['h_clusterid'] = cl.labels_
wine
```

```python
## K-measn clustering
from sklearn.cluster import KMeans
wcss = []
for i in range(1,6):
 kmeans = KMeans(n_clusters=i,random_state=2)
 kmeans.fit(dt4)
 wcss.append(kmeans.inertia_)
## WCSS Plot
plt.plot(range(1,6), wcss)
plt.title('Elbow Method')
plt.xlabel = ('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## Cluster algorithum
clusternew = KMeans(3,random_state=30).fit(dt4)
clusternew
```

```python
clusternew.labels_
```

```python
## Assigning Clusters to data set
wine2=dt.copy()
wine2['k_clustersid']=clusternew.labels_
wine2
```
